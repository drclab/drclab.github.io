```python
# --------------------------------------------------------------
# Revised cell ‚Äì produces JSON output
# --------------------------------------------------------------
from litellm import completion
import json

# 1Ô∏è‚É£  Call the model
response = completion(
    model="groq/openai/gpt-oss-120b",
    messages=[{"role": "user", "content": "hello from litellm"}],
)

# 2Ô∏è‚É£  Convert the Pydantic response object to a dict.
#     litellm uses Pydantic v2, so `model_dump()` is the canonical way.
#     (If you are on an older version, you can fall back to `dict()`.)
try:
    response_dict = response.model_dump()
except AttributeError:               # pragma: no cover
    # fallback for older litellm versions that use Pydantic v1
    response_dict = response.dict()

# 3Ô∏è‚É£  Print the dict as JSON (indented for readability)
print(json.dumps(response_dict, indent=2, ensure_ascii=False))
```
